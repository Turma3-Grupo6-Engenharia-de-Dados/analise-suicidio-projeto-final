## Hipóteses

1. H1 - Há um maior número de casos entre adolescentes (12 a 18 anos);
2. H2 - O número de casos apresenta um padrão sazonal, variando de acordo com o `MES` do ano, ou seja, datas comemorativas tedem ter maior número de ocorrências;
3. H3 - A distribuição de `IDADE` é diferente entre os grupos do `SEXO` masculino e feminino.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
#Carregar dados
suicidio=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/suicidioFinal.csv', encoding='latin-1', sep=",") #Dados do suicidio
```

- A base de dados já passou por um processo de limpeza anterior

```python
# Agrupar casos com mesmas caracteristicas
df_suicidio = suicidio.groupby(
    ['ESTADO', 'ANO', 'SEXO', 'RACACOR', 'IDADE', 'MES']
).size().reset_index(name='Qtde')

df_suicidio
```

```python
modelo_suicidio = sm.OLS.from_formula("Qtde ~ " \
"IDADE + " \
"ANO + " \
"SEXO + " \
"RACACOR   + " \
"MES + " \
"ESTADO", df_suicidio).fit()

modelo_suicidio.summary()
```

- Ainda que o modelo apresente um R² baixo, o modelo é estatistica significativo dado o p-value das variáves independentes (SEXO, RACACOR, ESTADO, ANO e IDADE). Uma possivel justifica para esse R² baixo foi a perda de dados que poderiam explicar melhor a influencia sob a quantidade de casos.

- Conforme levantando em hipotese, ser dor sexo masculino influencia muito na contagem dos casos.

- O mês apresenta um coef. muito baixo, o que implica no seu baixo efeito sob a quantidade

- A idade apresenta efeito negativo, ou seja, quanto mais velho menor a quantidade de casos, ou seja, o numero predomina em pessoas mais novas.

- Além disso, o estado de SP foi o que apresentou maior coef. se comparado aos demais, ou seja, nele prevalece a maior quantidade de casos
